## 3. ***Data Wrangling***

### Data Wrangling Code

````python
import pandas as pd
import numpy as np

def perform_comprehensive_wrangling(data):
    """
    Function to clean and transform the Mental Health in Tech dataset.
    Includes Exception Handling and Production Grade Logic.
    """
    try:
        # 1. Create a deep copy to avoid modifying the original dataframe
        df_cleaned = data.copy()

        # 2. Cleaning 'Age' column
        # Logic: Replace extreme outliers (like -1726 or 99999999) with the median of a realistic range (18-75)
        realistic_age_mask = (df_cleaned['Age'] >= 18) & (df_cleaned['Age'] <= 75)
        median_val = df_cleaned.loc[realistic_age_mask, 'Age'].median()
        df_cleaned['Age'] = df_cleaned['Age'].apply(lambda x: x if 18 <= x <= 75 else median_val)

        # 3. Standardizing 'Gender'
        # Logic: Using a mapping approach to collapse 40+ unique strings into 3 distinct categories
        def gender_standardizer(gender):
            gender = str(gender).lower().strip()
            male_keywords = ['m', 'male', 'cis male', 'man', 'mail', 'male-ish', 'maile', 'mal', 'male (cis)', 'make', 'cis man', 'msle', 'malr']
            female_keywords = ['f', 'female', 'cis female', 'woman', 'femake', 'femail', 'cis-female/femme', 'female (cis)', 'female ']

            if gender in male_keywords:
                return 'Male'
            elif gender in female_keywords:
                return 'Female'
            else:
                return 'Other/Non-binary'

        df_cleaned['Gender'] = df_cleaned['Gender'].apply(gender_standardizer)

        # 4. Imputing Missing Values
        # self_employed: Filling missing values with 'No' (Mode)
        df_cleaned['self_employed'] = df_cleaned['self_employed'].fillna('No')

        # work_interfere: Filling missing values with 'Unknown' as these are non-responders
        df_cleaned['work_interfere'] = df_cleaned['work_interfere'].fillna('Unknown')

        # 5. Dropping non-essential columns
        # Logic: Timestamp is not useful for behavior analysis; comments and state are 50%+ empty
        cols_to_remove = ['Timestamp', 'comments', 'state']
        df_cleaned.drop(columns=[c for c in cols_to_remove if c in df_cleaned.columns], inplace=True)

        # 6. Feature Engineering (For Multivariate Analysis)
        # Creating a numeric 'treatment_numeric' column for correlation purposes
        df_cleaned['treatment_numeric'] = df_cleaned['treatment'].map({'Yes': 1, 'No': 0})

        print("Data Wrangling complete: Errors handled and logic applied successfully.")
        return df_cleaned

    except KeyError as e:
        print(f"Production Error: Missing expected column in dataset: {e}")
        return data
    except Exception as e:
        print(f"Unexpected Error during Wrangling: {e}")
        return data

# Execute the Wrangling function
df = perform_comprehensive_wrangling(df)

# Final check of the wrangled data
df.head()
````

### What all manipulations have you done and insights you found?


**Data Manipulations Performed:**

Outlier Normalization: Identified and capped the 'Age' variable within a 18-75 range using median imputation to prevent biased statistical results.

Categorical Consolidation: Handled the high-cardinality 'Gender' column by implementing a keyword-based mapping logic to reduce noise and enable meaningful bivariate comparisons.

Null Imputation Strategy: Applied a non-destructive imputation for 'self_employed' (using the mode) and 'work_interfere' (using a new 'Unknown' category) to preserve row count while acknowledging data gaps.

Dimensionality Reduction: Pruned high-nullity columns (state, comments) and time-based noise (Timestamp) to focus the analysis on workplace behavior variables.

Feature Encoding: Created a numeric mapping for the 'treatment' variable to facilitate heatmaps and correlation analysis later in the notebook.


**Insights Found During Wrangling:**

Age Quality: The raw data contained significant entry errors (e.g., ages like -1726), suggesting the survey form lacked validation.

Diversity: The gender standardizing revealed a small but distinct percentage of non-binary/other respondents, showing the need for inclusive mental health policies.

Response Patterns: The large number of nulls in 'work_interfere' suggests that many employees may not be comfortable diagnosing their own productivity loss or are unsure of the signs.
